# Exploratory Data Analysis: S&P 500

This notebook gives a lightweight review of the processed S&P 500 dataset used in this project. It is meant to show exploratory thinking, not replace the main pipeline scripts.

## 1. Import Libraries and Load Data

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("seaborn-v0_8")

data_path = Path("../data/processed/sp500_processed.csv")
df = pd.read_csv(data_path, parse_dates=["Date"], index_col="Date")

df.head()
```

## 2. Quick Data Inspection

```python
df.info()
```

```python
df.describe()
```

## 3. Closing Price Trend

```python
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["Close"], color="navy")
plt.title("S&P 500 Closing Price")
plt.xlabel("Date")
plt.ylabel("Index Level")
plt.grid(True, alpha=0.3)
plt.show()
```

**Observation:** The index trends upward over the long run, but there are clear periods of market stress and recovery.

## 4. Daily Return Behavior

```python
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["Daily_Return"], color="darkgreen", linewidth=0.9)
plt.axhline(0, color="black", linestyle="--", linewidth=1)
plt.title("S&P 500 Daily Returns")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.grid(True, alpha=0.3)
plt.show()
```

**Observation:** Most daily returns stay near zero, but large spikes appear during high-volatility periods.

## 5. Rolling Volatility

```python
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["Rolling_Volatility"], color="firebrick")
plt.title("S&P 500 20-Day Rolling Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.grid(True, alpha=0.3)
plt.show()
```

**Observation:** Volatility increases during turbulent market periods, making it a useful short-term risk measure.

## 6. Key Takeaways

- The S&P 500 shows strong long-term growth, but the path is uneven.
- Daily returns are usually small, while extreme days drive a meaningful share of risk.
- Rolling volatility helps highlight periods of elevated market uncertainty.
- The processed dataset is ready for further analysis, dashboards, or modeling.
